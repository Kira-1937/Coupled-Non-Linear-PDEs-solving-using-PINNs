```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import time
import math
from tqdm import tqdm

# Check if CUDA is available and set the device accordingly
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

class PINN(nn.Module):
    def __init__(self, hidden_layers, neurons_per_layer):
        super(PINN, self).__init__()

        # Input layer (x, t) -> 2 inputs
        layers = [nn.Linear(2, neurons_per_layer), nn.Tanh()]

        # Hidden layers
        for _ in range(hidden_layers):
            layers.append(nn.Linear(neurons_per_layer, neurons_per_layer))
            layers.append(nn.Tanh())

        # Output layer for u and v (2 outputs)
        layers.append(nn.Linear(neurons_per_layer, 2))

        self.network = nn.Sequential(*layers)

    def forward(self, x, t):
        # Combine inputs into a single tensor
        inputs = torch.cat([x, t], dim=1)
        outputs = self.network(inputs)
        return outputs

    def u(self, x, t):
        outputs = self.forward(x, t)
        return outputs[:, 0:1]

    def v(self, x, t):
        outputs = self.forward(x, t)
        return outputs[:, 1:2]

    def u_derivatives(self, x, t):
        x.requires_grad_(True)
        t.requires_grad_(True)

        u = self.u(x, t)

        # First derivatives
        u_x = torch.autograd.grad(
            u, x, grad_outputs=torch.ones_like(u),
            create_graph=True, retain_graph=True
        )[0]

        u_t = torch.autograd.grad(
            u, t, grad_outputs=torch.ones_like(u),
            create_graph=True, retain_graph=True
        )[0]

        # Second derivatives
        u_xx = torch.autograd.grad(
            u_x, x, grad_outputs=torch.ones_like(u_x),
            create_graph=True, retain_graph=True
        )[0]

        return u, u_t, u_x, u_xx

    def v_derivatives(self, x, t):
        x.requires_grad_(True)
        t.requires_grad_(True)

        v = self.v(x, t)

        # First derivatives
        v_x = torch.autograd.grad(
            v, x, grad_outputs=torch.ones_like(v),
            create_graph=True, retain_graph=True
        )[0]

        v_t = torch.autograd.grad(
            v, t, grad_outputs=torch.ones_like(v),
            create_graph=True, retain_graph=True
        )[0]

        # Second derivatives
        v_xx = torch.autograd.grad(
            v_x, x, grad_outputs=torch.ones_like(v_x),
            create_graph=True, retain_graph=True
        )[0]

        return v, v_t, v_x, v_xx

    def compute_pde_residuals(self, x, t):
        # Get derivatives for u
        u, u_t, u_x, u_xx = self.u_derivatives(x, t)

        # Get derivatives for v
        v, v_t, v_x, v_xx = self.v_derivatives(x, t)

        # Calculate (u*v)_x term
        uv_x = torch.autograd.grad(
            u * v, x, grad_outputs=torch.ones_like(u),
            create_graph=True, retain_graph=True
        )[0]

        # Compute PDE residuals for Problem 2
        # u_t - u_xx - 2*u*u_x + 0.1*(u*v)_x = 0
        f_u = u_t - u_xx - 2 * u * u_x + 0.1 * uv_x

        # v_t - v_xx - 2*v*v_x + 0.3*(u*v)_x = 0
        f_v = v_t - v_xx - 2 * v * v_x + 0.3 * uv_x

        return f_u, f_v


def exact_solution(x, t):
    """
    Exact solution for the coupled Burgers equation (Problem 2):
    u(x, t) = 0.05 * (1 - tanh(-0.00625 * (x + 2*0.00625 * t)))
    v(x, t) = 0.05 * (-0.5 - tanh(-0.00625 * (x + 2*0.00625 * t)))
    """
    # Constants as defined in the problem
    k = 0.00625

    # Compute the argument for tanh
    arg = -k * (x + 2*k * t)

    # Compute u and v according to the exact solution
    u = 0.05 * (1 - torch.tanh(arg))
    v = 0.05 * (-0.5 - torch.tanh(arg))

    return u, v


def compute_errors(model, x_domain, t_values):
    """Compute L2 and L_infinity errors at specified time points"""
    model.eval()
    errors = {}

    for t_val in t_values:
        # Create mesh grid for this time value
        t_tensor = torch.full_like(x_domain, t_val).to(device)

        # Predict with model
        with torch.no_grad():
            u_pred = model.u(x_domain, t_tensor).cpu()
            v_pred = model.v(x_domain, t_tensor).cpu()

        # Calculate exact solution
        u_exact, v_exact = exact_solution(x_domain.cpu(), t_tensor.cpu())

        # Compute errors
        u_l2 = torch.sqrt(torch.mean((u_pred - u_exact)**2)).item()
        u_linf = torch.max(torch.abs(u_pred - u_exact)).item()

        v_l2 = torch.sqrt(torch.mean((v_pred - v_exact)**2)).item()
        v_linf = torch.max(torch.abs(v_pred - v_exact)).item()

        errors[t_val] = {
            'u_l2': u_l2,
            'u_linf': u_linf,
            'v_l2': v_l2,
            'v_linf': v_linf
        }

    return errors


def plot_3d_solutions(model, domain_limits, resolution=100):
    """Create 3D plot of numerical solution for u and v"""
    x_min, x_max = domain_limits[0]
    t_min, t_max = domain_limits[1]

    # Create mesh grid
    x = np.linspace(x_min, x_max, resolution)
    t = np.linspace(t_min, t_max, resolution)
    X, T = np.meshgrid(x, t)

    # Prepare input for model
    x_flat = torch.tensor(X.flatten(), dtype=torch.float32).reshape(-1, 1).to(device)
    t_flat = torch.tensor(T.flatten(), dtype=torch.float32).reshape(-1, 1).to(device)

    # Predict with model
    with torch.no_grad():
        u_pred = model.u(x_flat, t_flat).cpu().numpy().reshape(X.shape)
        v_pred = model.v(x_flat, t_flat).cpu().numpy().reshape(X.shape)

    # Create 3D plots
    fig = plt.figure(figsize=(16, 6))

    # Plot for u
    ax1 = fig.add_subplot(121, projection='3d')
    surf1 = ax1.plot_surface(X, T, u_pred, cmap=cm.viridis, linewidth=0, antialiased=False)
    ax1.set_xlabel('x')
    ax1.set_ylabel('t')
    ax1.set_zlabel('u(x,t)')
    ax1.set_title('3D numerical solution for u - Problem 2')
    fig.colorbar(surf1, ax=ax1, shrink=0.5, aspect=5)

    # Plot for v
    ax2 = fig.add_subplot(122, projection='3d')
    surf2 = ax2.plot_surface(X, T, v_pred, cmap=cm.viridis, linewidth=0, antialiased=False)
    ax2.set_xlabel('x')
    ax2.set_ylabel('t')
    ax2.set_zlabel('v(x,t)')
    ax2.set_title('3D numerical solution for v - Problem 2')
    fig.colorbar(surf2, ax=ax2, shrink=0.5, aspect=5)

    plt.tight_layout()
    plt.savefig('3d_solution_problem2.png', dpi=300, bbox_inches='tight')
    plt.show()


def generate_training_points(domain_limits, n_collocation_points=10000,
                             n_initial_points=200, n_boundary_points=200):
    """Generate training points for the PINN"""
    x_min, x_max = domain_limits[0]
    t_min, t_max = domain_limits[1]

    # Generate collocation points for enforcing PDE
    x_collocation = torch.FloatTensor(n_collocation_points, 1).uniform_(x_min, x_max).to(device)
    t_collocation = torch.FloatTensor(n_collocation_points, 1).uniform_(t_min, t_max).to(device)

    # Generate points for initial condition at t=0
    x_initial = torch.FloatTensor(n_initial_points, 1).uniform_(x_min, x_max).to(device)
    t_initial = torch.zeros(n_initial_points, 1).to(device)

    # Generate points for boundary conditions at x=-10 and x=10
    t_boundary = torch.FloatTensor(n_boundary_points, 1).uniform_(t_min, t_max).to(device)
    x_boundary_left = torch.full((n_boundary_points//2, 1), x_min).to(device)
    x_boundary_right = torch.full((n_boundary_points//2, 1), x_max).to(device)
    x_boundary = torch.cat([x_boundary_left, x_boundary_right], dim=0)
    t_boundary = torch.cat([t_boundary[:n_boundary_points//2], t_boundary[n_boundary_points//2:]], dim=0)

    return x_collocation, t_collocation, x_initial, t_initial, x_boundary, t_boundary


def find_best_hyperparameters(domain_limits, hidden_layers_range, neurons_range,
                             train_epochs=1000, eval_time=0.5):
    """
    Find the best hyperparameters (hidden_layers, neurons) by training multiple models
    and evaluating their performance
    """
    print("\nPerforming hyperparameter search...")
    results = []

    # Generate training points (same for all models)
    x_collocation, t_collocation, x_initial, t_initial, x_boundary, t_boundary = generate_training_points(domain_limits)

    # Calculate exact solutions for initial and boundary conditions
    u_exact_init, v_exact_init = exact_solution(x_initial, t_initial)
    u_exact_bound, v_exact_bound = exact_solution(x_boundary, t_boundary)

    # Create test domain for evaluation
    x_eval = torch.linspace(-10, 10, 100).reshape(-1, 1).to(device)

    best_error = float('inf')
    best_config = None

    # Create a grid of hyperparameter combinations
    configs = []
    for hidden in hidden_layers_range:
        for neurons in neurons_range:
            configs.append((hidden, neurons))

    total_configs = len(configs)
    print(f"Testing {total_configs} hyperparameter configurations...")

    # Loop through all hyperparameter combinations
    for i, (hidden, neurons) in enumerate(configs):
        print(f"\nConfiguration {i+1}/{total_configs}: hidden_layers={hidden}, neurons={neurons}")

        # Initialize model with current hyperparameters
        model = PINN(hidden_layers=hidden, neurons_per_layer=neurons).to(device)
        optimizer = optim.Adam(model.parameters(), lr=0.001)

        # Train for a fixed number of epochs
        pbar = tqdm(range(train_epochs))
        for epoch in pbar:
            optimizer.zero_grad()

            # PDE loss
            f_u, f_v = model.compute_pde_residuals(x_collocation, t_collocation)
            pde_loss = torch.mean(f_u**2) + torch.mean(f_v**2)

            # Initial condition loss
            u_pred_init = model.u(x_initial, t_initial)
            v_pred_init = model.v(x_initial, t_initial)
            ic_loss = torch.mean((u_pred_init - u_exact_init)**2) + torch.mean((v_pred_init - v_exact_init)**2)

            # Boundary condition loss
            u_pred_bound = model.u(x_boundary, t_boundary)
            v_pred_bound = model.v(x_boundary, t_boundary)
            bc_loss = torch.mean((u_pred_bound - u_exact_bound)**2) + torch.mean((v_pred_bound - v_exact_bound)**2)

            # Total loss
            loss = pde_loss + 10 * ic_loss + 10 * bc_loss
            loss.backward()
            optimizer.step()

            if epoch % 100 == 0:
                pbar.set_description(f"Loss: {loss.item():.6e}")

        # Evaluate model at t=eval_time
        errors = compute_errors(model, x_eval, [eval_time])
        current_error = errors[eval_time]['u_linf'] + errors[eval_time]['v_linf']  # Combined error

        results.append({
            'hidden_layers': hidden,
            'neurons': neurons,
            'u_linf_error': errors[eval_time]['u_linf'],
            'v_linf_error': errors[eval_time]['v_linf'],
            'combined_error': current_error
        })

        print(f"L∞ Error at t={eval_time}: u={errors[eval_time]['u_linf']:.6e}, v={errors[eval_time]['v_linf']:.6e}")

        # Update best configuration if current is better
        if current_error < best_error:
            best_error = current_error
            best_config = (hidden, neurons)
            print(f"New best configuration found!")

    # Sort results by combined error
    results.sort(key=lambda x: x['combined_error'])

    # Plot results
    plt.figure(figsize=(12, 8))

    # Create a scatter plot of hidden_layers vs neurons, with point size based on error (smaller error = larger point)
    hidden_values = [r['hidden_layers'] for r in results]
    neuron_values = [r['neurons'] for r in results]
    error_values = [1 / (r['combined_error'] + 1e-10) * 1000 for r in results]  # Inverse for visualization

    scatter = plt.scatter(hidden_values, neuron_values, s=error_values,
                         c=[r['combined_error'] for r in results], cmap='viridis_r', alpha=0.7)

    plt.colorbar(scatter, label='Combined Error')
    plt.xlabel('Number of Hidden Layers')
    plt.ylabel('Neurons per Layer')
    plt.title('Hyperparameter Search Results')

    # Annotate best point
    best_idx = [i for i, r in enumerate(results) if r['hidden_layers'] == best_config[0] and r['neurons'] == best_config[1]][0]
    plt.annotate('Best', (hidden_values[best_idx], neuron_values[best_idx]),
                xytext=(10, 10), textcoords='offset points',
                arrowprops=dict(arrowstyle='->'))

    plt.tight_layout()
    plt.savefig('hyperparameter_search_results.png', dpi=300, bbox_inches='tight')
    plt.show()

    # Print results table
    print("\nHyperparameter Search Results (Sorted by Error):")
    print("=" * 75)
    print(f"{'Rank':<5} | {'Hidden Layers':<15} | {'Neurons':<10} | {'u L∞ Error':<15} | {'v L∞ Error':<15}")
    print("-" * 75)

    for i, result in enumerate(results[:10]):  # Show top 10
        print(f"{i+1:<5} | {result['hidden_layers']:<15} | {result['neurons']:<10} | "
              f"{result['u_linf_error']:<15.6e} | {result['v_linf_error']:<15.6e}")

    print("=" * 75)
    print(f"Best configuration: hidden_layers={best_config[0]}, neurons={best_config[1]}")

    return best_config


def train_model(model, optimizer, domain_limits, n_collocation_points=10000,
               n_initial_points=200, n_boundary_points=200, epochs=10000,
               verbose=True, check_interval=500):
    """
    Train the PINN model to solve the coupled Burgers equation for Problem 2
    """
    # Generate training points
    x_collocation, t_collocation, x_initial, t_initial, x_boundary, t_boundary = generate_training_points(
        domain_limits, n_collocation_points, n_initial_points, n_boundary_points
    )

    # Calculate exact solutions for initial and boundary conditions
    u_exact_init, v_exact_init = exact_solution(x_initial, t_initial)
    u_exact_bound, v_exact_bound = exact_solution(x_boundary, t_boundary)

    # Training loop
    start_time = time.time()
    history = {'loss': [], 'pde_loss': [], 'ic_loss': [], 'bc_loss': [], 'time': []}

    pbar = tqdm(range(epochs)) if verbose else range(epochs)
    for epoch in pbar:
        optimizer.zero_grad()

        # PDE loss for collocation points
        f_u, f_v = model.compute_pde_residuals(x_collocation, t_collocation)
        pde_loss = torch.mean(f_u**2) + torch.mean(f_v**2)

        # Initial condition loss
        u_pred_init = model.u(x_initial, t_initial)
        v_pred_init = model.v(x_initial, t_initial)
        ic_loss = torch.mean((u_pred_init - u_exact_init)**2) + torch.mean((v_pred_init - v_exact_init)**2)

        # Boundary condition loss
        u_pred_bound = model.u(x_boundary, t_boundary)
        v_pred_bound = model.v(x_boundary, t_boundary)
        bc_loss = torch.mean((u_pred_bound - u_exact_bound)**2) + torch.mean((v_pred_bound - v_exact_bound)**2)

        # Total loss with weighting factors
        loss = pde_loss + 10 * ic_loss + 10 * bc_loss

        loss.backward()
        optimizer.step()

        # Store loss history
        if epoch % check_interval == 0:
            elapsed = time.time() - start_time
            history['loss'].append(loss.item())
            history['pde_loss'].append(pde_loss.item())
            history['ic_loss'].append(ic_loss.item())
            history['bc_loss'].append(bc_loss.item())
            history['time'].append(elapsed)

            if verbose:
                pbar.set_description(f"Loss: {loss.item():.6e}")

    # Plot loss history
    plt.figure(figsize=(12, 6))
    plt.semilogy(range(0, epochs, check_interval), history['loss'], label='Total Loss')
    plt.semilogy(range(0, epochs, check_interval), history['pde_loss'], label='PDE Loss')
    plt.semilogy(range(0, epochs, check_interval), history['ic_loss'], label='IC Loss')
    plt.semilogy(range(0, epochs, check_interval), history['bc_loss'], label='BC Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.title('Training Loss History - Problem 2')
    plt.savefig('loss_history_problem2.png', dpi=300, bbox_inches='tight')

    total_time = time.time() - start_time
    print(f"Training completed in {total_time:.2f} seconds")

    return model


def compare_solutions(model, t_values=[0.5, 1.0, 5.0, 10.0]):
    """Compare PINN predictions with exact solutions at specific times for Problem 2"""
    x = torch.linspace(-10, 10, 200).reshape(-1, 1).to(device)

    # Set up the figure
    fig, axes = plt.subplots(len(t_values), 2, figsize=(14, 4*len(t_values)))

    for i, t_val in enumerate(t_values):
        # Create time tensor
        t = torch.full_like(x, t_val).to(device)

        # Get PINN predictions
        with torch.no_grad():
            u_pred = model.u(x, t).cpu().numpy()
            v_pred = model.v(x, t).cpu().numpy()

        # Calculate exact solutions
        u_exact, v_exact = exact_solution(x.cpu(), torch.tensor([[t_val]]).cpu())
        u_exact = u_exact.numpy()
        v_exact = v_exact.numpy()

        # Plot u solution
        if len(t_values) == 1:
            ax_u = axes[0]
            ax_v = axes[1]
        else:
            ax_u = axes[i, 0]
            ax_v = axes[i, 1]

        ax_u.plot(x.cpu().numpy(), u_pred, 'r-', label='PINN')
        ax_u.plot(x.cpu().numpy(), u_exact, 'b--', label='Exact')
        ax_u.set_xlabel('x')
        ax_u.set_ylabel('u(x,t)')
        ax_u.set_title(f'u at t = {t_val}')
        ax_u.legend()
        ax_u.grid(True)

        # Plot v solution
        ax_v.plot(x.cpu().numpy(), v_pred, 'r-', label='PINN')
        ax_v.plot(x.cpu().numpy(), v_exact, 'b--', label='Exact')
        ax_v.set_xlabel('x')
        ax_v.set_ylabel('v(x,t)')
        ax_v.set_title(f'v at t = {t_val}')
        ax_v.legend()
        ax_v.grid(True)

    plt.tight_layout()
    plt.savefig('solution_comparison_problem2.png', dpi=300, bbox_inches='tight')
    plt.show()


def main():
    """Main function to run the PINN solution for Problem 2"""
    torch.manual_seed(42)  # For reproducibility

    # Define domain limits for Problem 2
    domain_limits = [(-10, 10), (0, 10)]  # [x_domain, t_domain]

    # First, find the best hyperparameters
    # Define the ranges for hyperparameter search
    hidden_layers_range = [4, 5, 6, 7, 8]
    neurons_range = [20, 30, 40, 50, 60]

    # Run hyperparameter search (with fewer epochs for quick search)
    best_hidden_layers, best_neurons = find_best_hyperparameters(
        domain_limits,
        hidden_layers_range,
        neurons_range,
        train_epochs=1000  # Fewer epochs for quick search
    )

    print(f"\nTraining final model with best hyperparameters: hidden_layers={best_hidden_layers}, neurons={best_neurons}")

    # Initialize the model with best hyperparameters
    model = PINN(hidden_layers=best_hidden_layers, neurons_per_layer=best_neurons).to(device)
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    # Print model architecture
    print(model)

    # Train the model with best hyperparameters
    print("Training final model for Problem 2...")
    model = train_model(
        model, optimizer, domain_limits, epochs=5000, check_interval=500
    )

    # Evaluate model performance at specific times
    print("\nCalculating errors for specific time points...")
    x_domain = torch.linspace(-10, 10, 200).reshape(-1, 1).to(device)
    errors = compute_errors(model, x_domain, [0.5, 1.0, 5.0, 10.0])

    # Print error metrics in a table format
    print("\nError metrics for Problem 2:")
    print("=" * 60)
    print(f"{'t':<5} | {'u L∞':<15} | {'u L2':<15} | {'v L∞':<15} | {'v L2':<15}")
    print("-" * 60)
    for t_val, err in errors.items():
        print(f"{t_val:<5} | {err['u_linf']:<15.6e} | {err['u_l2']:<15.6e} | {err['v_linf']:<15.6e} | {err['v_l2']:<15.6e}")
    print("=" * 60)

    # Create plots
    print("\nGenerating solution plots...")
    compare_solutions(model)
    plot_3d_solutions(model, domain_limits)

    print("\nAnalysis for Problem 2 complete!")


if __name__ == "__main__":
    main()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import time
import math
from tqdm import tqdm

# Check if CUDA is available and set the device accordingly
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

class PINN(nn.Module):
    def __init__(self, hidden_layers, neurons_per_layer):
        super(PINN, self).__init__()

        # Input layer (x, t) -> 2 inputs
        layers = [nn.Linear(2, neurons_per_layer), nn.Tanh()]

        # Hidden layers
        for _ in range(hidden_layers):
            layers.append(nn.Linear(neurons_per_layer, neurons_per_layer))
            layers.append(nn.Tanh())

        # Output layer for u and v (2 outputs)
        layers.append(nn.Linear(neurons_per_layer, 2))

        self.network = nn.Sequential(*layers)

    def forward(self, x, t):
        # Combine inputs into a single tensor
        inputs = torch.cat([x, t], dim=1)
        outputs = self.network(inputs)
        return outputs

    def u(self, x, t):
        outputs = self.forward(x, t)
        return outputs[:, 0:1]

    def v(self, x, t):
        outputs = self.forward(x, t)
        return outputs[:, 1:2]

    def u_derivatives(self, x, t):
        x.requires_grad_(True)
        t.requires_grad_(True)

        u = self.u(x, t)

        # First derivatives
        u_x = torch.autograd.grad(
            u, x, grad_outputs=torch.ones_like(u),
            create_graph=True, retain_graph=True
        )[0]

        u_t = torch.autograd.grad(
            u, t, grad_outputs=torch.ones_like(u),
            create_graph=True, retain_graph=True
        )[0]

        # Second derivatives
        u_xx = torch.autograd.grad(
            u_x, x, grad_outputs=torch.ones_like(u_x),
            create_graph=True, retain_graph=True
        )[0]

        return u, u_t, u_x, u_xx

    def v_derivatives(self, x, t):
        x.requires_grad_(True)
        t.requires_grad_(True)

        v = self.v(x, t)

        # First derivatives
        v_x = torch.autograd.grad(
            v, x, grad_outputs=torch.ones_like(v),
            create_graph=True, retain_graph=True
        )[0]

        v_t = torch.autograd.grad(
            v, t, grad_outputs=torch.ones_like(v),
            create_graph=True, retain_graph=True
        )[0]

        # Second derivatives
        v_xx = torch.autograd.grad(
            v_x, x, grad_outputs=torch.ones_like(v_x),
            create_graph=True, retain_graph=True
        )[0]

        return v, v_t, v_x, v_xx

    def compute_pde_residuals(self, x, t, epsilon=1e-6):
        """
        Compute PDE residuals for Problem 3:
        u_t - ε*u_xx + 2*u*u_x - (u*v)_x = 0
        v_t - ε*v_xx + 2*v*v_x - (u*v)_x = 0
        """
        # Get derivatives for u
        u, u_t, u_x, u_xx = self.u_derivatives(x, t)

        # Get derivatives for v
        v, v_t, v_x, v_xx = self.v_derivatives(x, t)

        # Calculate (u*v)_x term
        uv_x = torch.autograd.grad(
            u * v, x, grad_outputs=torch.ones_like(u),
            create_graph=True, retain_graph=True
        )[0]

        # Compute PDE residuals for Problem 3
        # u_t - ε*u_xx + 2*u*u_x - (u*v)_x = 0
        f_u = u_t - epsilon * u_xx + 2 * u * u_x - uv_x

        # v_t - ε*v_xx + 2*v*v_x - (u*v)_x = 0
        f_v = v_t - epsilon * v_xx + 2 * v * v_x - uv_x

        return f_u, f_v


def exact_solution(x, t, epsilon=1e-6):
    """
    Exact solution for Problem 3:
    u(x, t) = v(x, t) = exp(-επ²t) * cos(πx)
    """
    # Calculate exact solutions
    factor = torch.exp(-epsilon * (math.pi**2) * t)
    solution = factor * torch.cos(math.pi * x)

    return solution, solution  # Return the same for both u and v


def compute_errors(model, x_domain, t_values, epsilon=1e-6):
    """Compute L2 and L_infinity errors at specified time points"""
    model.eval()
    errors = {}

    for t_val in t_values:
        # Create mesh grid for this time value
        t_tensor = torch.full_like(x_domain, t_val).to(device)

        # Predict with model
        with torch.no_grad():
            u_pred = model.u(x_domain, t_tensor).cpu()
            v_pred = model.v(x_domain, t_tensor).cpu()

        # Calculate exact solution
        u_exact, v_exact = exact_solution(x_domain.cpu(), t_tensor.cpu(), epsilon)

        # Compute errors
        u_l2 = torch.sqrt(torch.mean((u_pred - u_exact)**2)).item()
        u_linf = torch.max(torch.abs(u_pred - u_exact)).item()

        v_l2 = torch.sqrt(torch.mean((v_pred - v_exact)**2)).item()
        v_linf = torch.max(torch.abs(v_pred - v_exact)).item()

        errors[t_val] = {
            'u_l2': u_l2,
            'u_linf': u_linf,
            'v_l2': v_l2,
            'v_linf': v_linf
        }

    return errors


def plot_3d_solutions(model, domain_limits, epsilon=1e-6, resolution=100):
    """Create 3D plot of numerical solution for u and v"""
    x_min, x_max = domain_limits[0]
    t_min, t_max = domain_limits[1]

    # Create mesh grid
    x = np.linspace(x_min, x_max, resolution)
    t = np.linspace(t_min, t_max, resolution)
    X, T = np.meshgrid(x, t)

    # Prepare input for model
    x_flat = torch.tensor(X.flatten(), dtype=torch.float32).reshape(-1, 1).to(device)
    t_flat = torch.tensor(T.flatten(), dtype=torch.float32).reshape(-1, 1).to(device)

    # Predict with model
    with torch.no_grad():
        u_pred = model.u(x_flat, t_flat).cpu().numpy().reshape(X.shape)
        v_pred = model.v(x_flat, t_flat).cpu().numpy().reshape(X.shape)

    # Create 3D plots
    fig = plt.figure(figsize=(16, 6))

    # Plot for u
    ax1 = fig.add_subplot(121, projection='3d')
    surf1 = ax1.plot_surface(X, T, u_pred, cmap=cm.viridis, linewidth=0, antialiased=False)
    ax1.set_xlabel('x')
    ax1.set_ylabel('t')
    ax1.set_zlabel('u(x,t)')
    ax1.set_title(f'3D numerical solution for u - Problem 3 (ε={epsilon})')
    fig.colorbar(surf1, ax=ax1, shrink=0.5, aspect=5)

    # Plot for v
    ax2 = fig.add_subplot(122, projection='3d')
    surf2 = ax2.plot_surface(X, T, v_pred, cmap=cm.viridis, linewidth=0, antialiased=False)
    ax2.set_xlabel('x')
    ax2.set_ylabel('t')
    ax2.set_zlabel('v(x,t)')
    ax2.set_title(f'3D numerical solution for v - Problem 3 (ε={epsilon})')
    fig.colorbar(surf2, ax=ax2, shrink=0.5, aspect=5)

    plt.tight_layout()
    plt.savefig(f'3d_solution_problem3_eps_{epsilon}.png', dpi=300, bbox_inches='tight')
    plt.show()


def generate_domain_points(domain_limits, n_collocation_points=10000, n_initial_points=200, n_boundary_points=200):
    """Generate all required points for training"""
    x_min, x_max = domain_limits[0]
    t_min, t_max = domain_limits[1]

    # Generate collocation points for enforcing PDE
    x_collocation = torch.FloatTensor(n_collocation_points, 1).uniform_(x_min, x_max).to(device)
    t_collocation = torch.FloatTensor(n_collocation_points, 1).uniform_(t_min, t_max).to(device)

    # Generate points for initial condition at t=0
    x_initial = torch.FloatTensor(n_initial_points, 1).uniform_(x_min, x_max).to(device)
    t_initial = torch.zeros(n_initial_points, 1).to(device)

    # Generate points for boundary conditions at x=0 and x=1
    t_boundary = torch.FloatTensor(n_boundary_points, 1).uniform_(t_min, t_max).to(device)
    x_boundary_left = torch.full((n_boundary_points//2, 1), x_min).to(device)
    x_boundary_right = torch.full((n_boundary_points//2, 1), x_max).to(device)
    x_boundary = torch.cat([x_boundary_left, x_boundary_right], dim=0)
    t_boundary = torch.cat([t_boundary[:n_boundary_points//2], t_boundary[n_boundary_points//2:]], dim=0)

    return x_collocation, t_collocation, x_initial, t_initial, x_boundary, t_boundary


def hyperparameter_tuning(domain_limits, epsilon=1e-6, tuning_epochs=1000):
    """
    Perform hyperparameter tuning to find the best configuration
    """
    # Generate points for training
    x_collocation, t_collocation, x_initial, t_initial, x_boundary, t_boundary = generate_domain_points(domain_limits)

    # Calculate exact solutions for initial and boundary conditions
    u_exact_init, v_exact_init = exact_solution(x_initial, t_initial, epsilon)
    u_exact_bound, v_exact_bound = exact_solution(x_boundary, t_boundary, epsilon)

    # Hyperparameter grid
    hidden_layers_options = [3, 4, 5, 6, 7]
    neurons_options = [20, 30, 40, 50, 60]

    # Create test domain for evaluation
    x_test = torch.linspace(0, 1, 100).reshape(-1, 1).to(device)

    results = []
    best_error = float('inf')
    best_config = None

    print("Starting hyperparameter tuning...")
    print(f"{'Hidden Layers':<15} {'Neurons':<10} {'L∞ Error':<15} {'L2 Error':<15}")
    print("-" * 55)

    for hidden_layers in hidden_layers_options:
        for neurons in neurons_options:
            # Initialize model with current hyperparameters
            model = PINN(hidden_layers=hidden_layers, neurons_per_layer=neurons).to(device)
            optimizer = optim.Adam(model.parameters(), lr=0.001)

            # Short training cycle for tuning
            for epoch in range(tuning_epochs):
                optimizer.zero_grad()

                # PDE loss
                f_u, f_v = model.compute_pde_residuals(x_collocation, t_collocation, epsilon)
                pde_loss = torch.mean(f_u**2) + torch.mean(f_v**2)

                # Initial condition loss
                u_pred_init = model.u(x_initial, t_initial)
                v_pred_init = model.v(x_initial, t_initial)
                ic_loss = torch.mean((u_pred_init - u_exact_init)**2) + torch.mean((v_pred_init - v_exact_init)**2)

                # Boundary condition loss
                u_pred_bound = model.u(x_boundary, t_boundary)
                v_pred_bound = model.v(x_boundary, t_boundary)
                bc_loss = torch.mean((u_pred_bound - u_exact_bound)**2) + torch.mean((v_pred_bound - v_exact_bound)**2)

                # Total loss
                loss = pde_loss + 10 * ic_loss + 10 * bc_loss
                loss.backward()
                optimizer.step()

            # Evaluate model at t=0.5 (typical evaluation point)
            errors = compute_errors(model, x_test, [0.5], epsilon)
            u_l2_error = errors[0.5]['u_l2']
            u_linf_error = errors[0.5]['u_linf']

            # Store results
            config_result = {
                'hidden_layers': hidden_layers,
                'neurons': neurons,
                'u_l2_error': u_l2_error,
                'u_linf_error': u_linf_error
            }
            results.append(config_result)

            # Check if this is the best configuration so far
            if u_linf_error < best_error:
                best_error = u_linf_error
                best_config = {
                    'hidden_layers': hidden_layers,
                    'neurons': neurons
                }

            print(f"{hidden_layers:<15} {neurons:<10} {u_linf_error:<15.6e} {u_l2_error:<15.6e}")

    print("-" * 55)
    print(f"Best configuration: {best_config['hidden_layers']} hidden layers, {best_config['neurons']} neurons")
    print(f"Best L∞ error: {best_error:.6e}")

    # Plot hyperparameter tuning results as a heatmap
    plot_hyperparameter_results(results, hidden_layers_options, neurons_options)

    return best_config, results


def plot_hyperparameter_results(results, hidden_layers_options, neurons_options):
    """Create visualization of hyperparameter tuning results"""
    # Extract data for heatmap
    error_matrix = np.zeros((len(hidden_layers_options), len(neurons_options)))

    for result in results:
        hidden_idx = hidden_layers_options.index(result['hidden_layers'])
        neuron_idx = neurons_options.index(result['neurons'])
        error_matrix[hidden_idx, neuron_idx] = result['u_linf_error']

    # Create heatmap
    plt.figure(figsize=(10, 8))
    plt.imshow(error_matrix, cmap='viridis_r')  # Reverse colormap so darker is better

    # Add text annotations
    for i in range(len(hidden_layers_options)):
        for j in range(len(neurons_options)):
            plt.text(j, i, f"{error_matrix[i, j]:.2e}",
                     ha="center", va="center", color="w")

    # Set labels
    plt.colorbar(label='L∞ Error')
    plt.xlabel('Number of Neurons')
    plt.ylabel('Number of Hidden Layers')
    plt.xticks(range(len(neurons_options)), neurons_options)
    plt.yticks(range(len(hidden_layers_options)), hidden_layers_options)
    plt.title('Hyperparameter Tuning Results')

    plt.savefig(f'hyperparameter_tuning_heatmap.png', dpi=300, bbox_inches='tight')
    plt.show()


def train_model(model, optimizer, domain_limits, epsilon=1e-6, n_collocation_points=10000,
               n_initial_points=200, n_boundary_points=200, epochs=10000,
               verbose=True, check_interval=500):
    """
    Train the PINN model to solve the coupled equation for Problem 3
    """
    # Generate all required points
    x_collocation, t_collocation, x_initial, t_initial, x_boundary, t_boundary = generate_domain_points(
        domain_limits, n_collocation_points, n_initial_points, n_boundary_points
    )

    # Calculate exact solutions for initial and boundary conditions
    u_exact_init, v_exact_init = exact_solution(x_initial, t_initial, epsilon)
    u_exact_bound, v_exact_bound = exact_solution(x_boundary, t_boundary, epsilon)

    # Training loop
    start_time = time.time()
    history = {'loss': [], 'pde_loss': [], 'ic_loss': [], 'bc_loss': [], 'time': []}

    pbar = tqdm(range(epochs)) if verbose else range(epochs)
    for epoch in pbar:
        optimizer.zero_grad()

        # PDE loss for collocation points
        f_u, f_v = model.compute_pde_residuals(x_collocation, t_collocation, epsilon)
        pde_loss = torch.mean(f_u**2) + torch.mean(f_v**2)

        # Initial condition loss
        u_pred_init = model.u(x_initial, t_initial)
        v_pred_init = model.v(x_initial, t_initial)
        ic_loss = torch.mean((u_pred_init - u_exact_init)**2) + torch.mean((v_pred_init - v_exact_init)**2)

        # Boundary condition loss
        u_pred_bound = model.u(x_boundary, t_boundary)
        v_pred_bound = model.v(x_boundary, t_boundary)
        bc_loss = torch.mean((u_pred_bound - u_exact_bound)**2) + torch.mean((v_pred_bound - v_exact_bound)**2)

        # Total loss with weighting factors
        loss = pde_loss + 10 * ic_loss + 10 * bc_loss

        loss.backward()
        optimizer.step()

        # Store loss history
        if epoch % check_interval == 0:
            elapsed = time.time() - start_time
            history['loss'].append(loss.item())
            history['pde_loss'].append(pde_loss.item())
            history['ic_loss'].append(ic_loss.item())
            history['bc_loss'].append(bc_loss.item())
            history['time'].append(elapsed)

            if verbose:
                pbar.set_description(f"Loss: {loss.item():.6e}")

    # Plot loss history
    plt.figure(figsize=(12, 6))
    plt.semilogy(range(0, epochs, check_interval), history['loss'], label='Total Loss')
    plt.semilogy(range(0, epochs, check_interval), history['pde_loss'], label='PDE Loss')
    plt.semilogy(range(0, epochs, check_interval), history['ic_loss'], label='IC Loss')
    plt.semilogy(range(0, epochs, check_interval), history['bc_loss'], label='BC Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.title(f'Training Loss History - Problem 3 (ε={epsilon})')
    plt.savefig(f'loss_history_problem3_eps_{epsilon}.png', dpi=300, bbox_inches='tight')

    total_time = time.time() - start_time
    print(f"Training completed in {total_time:.2f} seconds")

    return model


def compare_solutions(model, epsilon=1e-6, t_values=[0.1, 0.3, 0.5, 0.7, 1.0]):
    """Compare PINN predictions with exact solutions at specific times for Problem 3"""
    x = torch.linspace(0, 1, 200).reshape(-1, 1).to(device)

    # Set up the figure
    fig, axes = plt.subplots(len(t_values), 2, figsize=(14, 4*len(t_values)))

    for i, t_val in enumerate(t_values):
        # Create time tensor
        t = torch.full_like(x, t_val).to(device)

        # Get PINN predictions
        with torch.no_grad():
            u_pred = model.u(x, t).cpu().numpy()
            v_pred = model.v(x, t).cpu().numpy()

        # Calculate exact solutions
        u_exact, v_exact = exact_solution(x.cpu(), torch.tensor([[t_val]]).cpu(), epsilon)
        u_exact = u_exact.numpy()
        v_exact = v_exact.numpy()

        # Plot u solution
        if len(t_values) == 1:
            ax_u = axes[0]
            ax_v = axes[1]
        else:
            ax_u = axes[i, 0]
            ax_v = axes[i, 1]

        ax_u.plot(x.cpu().numpy(), u_pred, 'r-', label='PINN')
        ax_u.plot(x.cpu().numpy(), u_exact, 'b--', label='Exact')
        ax_u.set_xlabel('x')
        ax_u.set_ylabel('u(x,t)')
        ax_u.set_title(f'u at t = {t_val}')
        ax_u.legend()
        ax_u.grid(True)

        # Plot v solution
        ax_v.plot(x.cpu().numpy(), v_pred, 'r-', label='PINN')
        ax_v.plot(x.cpu().numpy(), v_exact, 'b--', label='Exact')
        ax_v.set_xlabel('x')
        ax_v.set_ylabel('v(x,t)')
        ax_v.set_title(f'v at t = {t_val}')
        ax_v.legend()
        ax_v.grid(True)

    plt.tight_layout()
    plt.savefig(f'solution_comparison_problem3_eps_{epsilon}.png', dpi=300, bbox_inches='tight')
    plt.show()


def create_error_tables(model, epsilon=1e-6, t_values=[0.1, 0.3, 0.5, 0.7, 1.0]):
    """Create tables similar to Tables 7 and 8 in the paper"""
    x_domain = torch.linspace(0, 1, 200).reshape(-1, 1).to(device)
    errors = compute_errors(model, x_domain, t_values, epsilon)

    # Print error metrics in a table format for u
    print(f"\nErrors obtained for u with Re = {1/epsilon}, N = 4 for problem 3:")
    print("=" * 60)
    print(f"{'t':<5} | {'L∞':<15} | {'L2':<15}")
    print("-" * 60)
    for t_val, err in errors.items():
        print(f"{t_val:<5} | {err['u_linf']:<15.6e} | {err['u_l2']:<15.6e}")
    print("=" * 60)

    # Print error metrics in a table format for v
    print(f"\nErrors obtained for v with Re = {1/epsilon}, N = 4 for problem 3:")
    print("=" * 60)
    print(f"{'t':<5} | {'L∞':<15} | {'L2':<15}")
    print("-" * 60)
    for t_val, err in errors.items():
        print(f"{t_val:<5} | {err['v_linf']:<15.6e} | {err['v_l2']:<15.6e}")
    print("=" * 60)

    return errors


def main():
    """Main function to run the PINN solution for Problem 3"""
    torch.manual_seed(42)  # For reproducibility

    # Define domain limits for Problem 3
    domain_limits = [(0, 1), (0, 1)]  # [x_domain, t_domain]

    # Set Reynolds number (Re = 1/ε)
    Re = 1e6  # As specified in Table 7 and 8
    epsilon = 1.0 / Re
    print(f"Running Problem 3 with Re = {Re} (ε = {epsilon})")

    # STEP 1: Hyperparameter tuning
    print("\n=== HYPERPARAMETER TUNING ===")
    tuning_epochs = 1000  # Reduced epochs for tuning
    best_config, tuning_results = hyperparameter_tuning(domain_limits, epsilon, tuning_epochs)

    # Use the best configuration from hyperparameter tuning
    hidden_layers = best_config['hidden_layers']
    neurons = best_config['neurons']
    print(f"\nUsing optimal configuration with {hidden_layers} hidden layers and {neurons} neurons per layer")

    # STEP 2: Train the model with best hyperparameters
    print("\n=== TRAINING FINAL MODEL ===")
    learning_rate = 0.001
    epochs = 5000

    # Initialize the model with best hyperparameters
    model = PINN(hidden_layers=hidden_layers, neurons_per_layer=neurons).to(device)
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)

    # Print model architecture
    print(model)

    # Train the model
    model = train_model(
        model, optimizer, domain_limits, epsilon=epsilon, epochs=epochs, check_interval=500
    )

    # Create error tables like in the paper
    t_values = [0.1, 0.3, 0.5, 0.7, 1.0]  # Time points from Tables 7 and 8
    errors = create_error_tables(model, epsilon, t_values)

    # Create plots
    print("\nGenerating solution plots...")
    compare_solutions(model, epsilon, t_values)
    plot_3d_solutions(model, domain_limits, epsilon)

    print("\nAnalysis for Problem 3 complete!")


if __name__ == "__main__":
    main()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import time
import math
from tqdm import tqdm

# Check if CUDA is available and set the device accordingly
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

class PINN(nn.Module):
    def __init__(self, hidden_layers, neurons_per_layer):
        super(PINN, self).__init__()

        # Input layer (x, t) -> 2 inputs
        layers = [nn.Linear(2, neurons_per_layer), nn.Tanh()]

        # Hidden layers
        for _ in range(hidden_layers):
            layers.append(nn.Linear(neurons_per_layer, neurons_per_layer))
            layers.append(nn.Tanh())

        # Output layer for u and v (2 outputs)
        layers.append(nn.Linear(neurons_per_layer, 2))

        self.network = nn.Sequential(*layers)

    def forward(self, x, t):
        # Combine inputs into a single tensor
        inputs = torch.cat([x, t], dim=1)
        outputs = self.network(inputs)
        return outputs

    def u(self, x, t):
        outputs = self.forward(x, t)
        return outputs[:, 0:1]

    def v(self, x, t):
        outputs = self.forward(x, t)
        return outputs[:, 1:2]

    def u_derivatives(self, x, t):
        x.requires_grad_(True)
        t.requires_grad_(True)

        u = self.u(x, t)

        # First derivatives
        u_x = torch.autograd.grad(
            u, x, grad_outputs=torch.ones_like(u),
            create_graph=True, retain_graph=True
        )[0]

        u_t = torch.autograd.grad(
            u, t, grad_outputs=torch.ones_like(u),
            create_graph=True, retain_graph=True
        )[0]

        # Second derivatives
        u_xx = torch.autograd.grad(
            u_x, x, grad_outputs=torch.ones_like(u_x),
            create_graph=True, retain_graph=True
        )[0]

        return u, u_t, u_x, u_xx

    def v_derivatives(self, x, t):
        x.requires_grad_(True)
        t.requires_grad_(True)

        v = self.v(x, t)

        # First derivatives
        v_x = torch.autograd.grad(
            v, x, grad_outputs=torch.ones_like(v),
            create_graph=True, retain_graph=True
        )[0]

        v_t = torch.autograd.grad(
            v, t, grad_outputs=torch.ones_like(v),
            create_graph=True, retain_graph=True
        )[0]

        # Second derivatives
        v_xx = torch.autograd.grad(
            v_x, x, grad_outputs=torch.ones_like(v_x),
            create_graph=True, retain_graph=True
        )[0]

        return v, v_t, v_x, v_xx

    def compute_pde_residuals(self, x, t):
        # Get derivatives for u
        u, u_t, u_x, u_xx = self.u_derivatives(x, t)

        # Get derivatives for v
        v, v_t, v_x, v_xx = self.v_derivatives(x, t)

        # Calculate (u*v)_x term
        uv_x = torch.autograd.grad(
            u * v, x, grad_outputs=torch.ones_like(u),
            create_graph=True, retain_graph=True
        )[0]

        # Compute PDE residuals
        # u_t - u_xx - 2*u*u_x + (u*v)_x = 0
        f_u = u_t - u_xx - 2 * u * u_x + uv_x

        # v_t - v_xx - 2*v*v_x + (u*v)_x = 0
        f_v = v_t - v_xx - 2 * v * v_x + uv_x

        return f_u, f_v


def exact_solution(x, t):
    """
    Exact solution for the coupled Burgers equation:
    u(x, t) = v(x, t) = exp(-t) * sin(x)
    """
    u = torch.exp(-t) * torch.sin(x)
    v = torch.exp(-t) * torch.sin(x)
    return u, v


def compute_errors(model, x_domain, t_values):
    """Compute L2 and L_infinity errors at specified time points"""
    model.eval()
    errors = {}

    for t_val in t_values:
        # Create mesh grid for this time value
        t_tensor = torch.full_like(x_domain, t_val).to(device)

        # Predict with model
        with torch.no_grad():
            u_pred = model.u(x_domain, t_tensor).cpu()
            v_pred = model.v(x_domain, t_tensor).cpu()

        # Calculate exact solution
        u_exact, v_exact = exact_solution(x_domain.cpu(), t_tensor.cpu())

        # Compute errors
        u_l2 = torch.sqrt(torch.mean((u_pred - u_exact)**2)).item()
        u_linf = torch.max(torch.abs(u_pred - u_exact)).item()

        v_l2 = torch.sqrt(torch.mean((v_pred - v_exact)**2)).item()
        v_linf = torch.max(torch.abs(v_pred - v_exact)).item()

        errors[t_val] = {
            'u_l2': u_l2,
            'u_linf': u_linf,
            'v_l2': v_l2,
            'v_linf': v_linf
        }

    return errors


def plot_3d_solutions(model, domain_limits, resolution=100):
    """Create 3D plot of numerical solution for u and v"""
    x_min, x_max = domain_limits[0]
    t_min, t_max = domain_limits[1]

    # Create mesh grid
    x = np.linspace(x_min, x_max, resolution)
    t = np.linspace(t_min, t_max, resolution)
    X, T = np.meshgrid(x, t)

    # Prepare input for model
    x_flat = torch.tensor(X.flatten(), dtype=torch.float32).reshape(-1, 1).to(device)
    t_flat = torch.tensor(T.flatten(), dtype=torch.float32).reshape(-1, 1).to(device)

    # Predict with model
    with torch.no_grad():
        u_pred = model.u(x_flat, t_flat).cpu().numpy().reshape(X.shape)
        v_pred = model.v(x_flat, t_flat).cpu().numpy().reshape(X.shape)

    # Create 3D plots
    fig = plt.figure(figsize=(16, 6))

    # Plot for u
    ax1 = fig.add_subplot(121, projection='3d')
    surf1 = ax1.plot_surface(X, T, u_pred, cmap=cm.viridis, linewidth=0, antialiased=False)
    ax1.set_xlabel('x')
    ax1.set_ylabel('t')
    ax1.set_zlabel('u(x,t)')
    ax1.set_title('3D numerical solution for u')
    fig.colorbar(surf1, ax=ax1, shrink=0.5, aspect=5)

    # Plot for v
    ax2 = fig.add_subplot(122, projection='3d')
    surf2 = ax2.plot_surface(X, T, v_pred, cmap=cm.viridis, linewidth=0, antialiased=False)
    ax2.set_xlabel('x')
    ax2.set_ylabel('t')
    ax2.set_zlabel('v(x,t)')
    ax2.set_title('3D numerical solution for v')
    fig.colorbar(surf2, ax=ax2, shrink=0.5, aspect=5)

    plt.tight_layout()
    plt.savefig('3d_solution.png', dpi=300, bbox_inches='tight')
    plt.show()


def train_and_evaluate_model(hidden_layers, neurons_per_layer, domain_limits, n_iterations=1000, eval_at_t=0.5):
    """Train model with specified hyperparameters and return error metrics"""
    model = PINN(hidden_layers=hidden_layers, neurons_per_layer=neurons_per_layer).to(device)
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    x_min, x_max = domain_limits[0]
    t_min, t_max = domain_limits[1]

    # Generate collocation points for enforcing PDE
    n_collocation_points = 5000
    x_collocation = torch.FloatTensor(n_collocation_points, 1).uniform_(x_min, x_max).to(device)
    t_collocation = torch.FloatTensor(n_collocation_points, 1).uniform_(t_min, t_max).to(device)

    # Generate points for initial condition at t=0
    n_initial_points = 200
    x_initial = torch.FloatTensor(n_initial_points, 1).uniform_(x_min, x_max).to(device)
    t_initial = torch.zeros(n_initial_points, 1).to(device)

    # Generate points for boundary conditions at x=-π and x=π
    n_boundary_points = 200
    t_boundary = torch.FloatTensor(n_boundary_points, 1).uniform_(t_min, t_max).to(device)
    x_boundary_left = torch.full((n_boundary_points//2, 1), x_min).to(device)
    x_boundary_right = torch.full((n_boundary_points//2, 1), x_max).to(device)
    x_boundary = torch.cat([x_boundary_left, x_boundary_right], dim=0)
    t_boundary = torch.cat([t_boundary[:n_boundary_points//2], t_boundary[n_boundary_points//2:]], dim=0)

    # Calculate exact solutions for initial and boundary conditions
    u_exact_init, v_exact_init = exact_solution(x_initial, t_initial)
    u_exact_bound, v_exact_bound = exact_solution(x_boundary, t_boundary)

    # Training loop
    pbar = tqdm(range(n_iterations), desc=f"Training {hidden_layers}x{neurons_per_layer}")
    for _ in pbar:
        optimizer.zero_grad()

        # PDE loss for collocation points
        f_u, f_v = model.compute_pde_residuals(x_collocation, t_collocation)
        pde_loss = torch.mean(f_u**2) + torch.mean(f_v**2)

        # Initial condition loss
        u_pred_init = model.u(x_initial, t_initial)
        v_pred_init = model.v(x_initial, t_initial)
        ic_loss = torch.mean((u_pred_init - u_exact_init)**2) + torch.mean((v_pred_init - v_exact_init)**2)

        # Boundary condition loss
        u_pred_bound = model.u(x_boundary, t_boundary)
        v_pred_bound = model.v(x_boundary, t_boundary)
        bc_loss = torch.mean((u_pred_bound - u_exact_bound)**2) + torch.mean((v_pred_bound - v_exact_bound)**2)

        # Total loss with weighting factors
        loss = pde_loss + 10 * ic_loss + 10 * bc_loss

        # Update progress bar with loss
        pbar.set_postfix(loss=f"{loss.item():.6e}")

        loss.backward()
        optimizer.step()

    # Evaluate model
    x_eval = torch.linspace(-math.pi, math.pi, 100).reshape(-1, 1).to(device)
    errors = compute_errors(model, x_eval, [eval_at_t])

    return {
        'model': model,
        'hidden_layers': hidden_layers,
        'neurons': neurons_per_layer,
        'u_l2': errors[eval_at_t]['u_l2'],
        'u_linf': errors[eval_at_t]['u_linf'],
        'v_l2': errors[eval_at_t]['v_l2'],
        'v_linf': errors[eval_at_t]['v_linf'],
        'total_error': errors[eval_at_t]['u_l2'] + errors[eval_at_t]['v_l2']
    }


def hyperparameter_tuning(domain_limits, hidden_layers_range, neurons_range, n_iterations=1000):
    """
    Perform hyperparameter tuning to find optimal architecture
    """
    results = []

    print("\n=== Starting Hyperparameter Tuning ===")
    print(f"Testing combinations of hidden layers {hidden_layers_range} and neurons {neurons_range}")
    print(f"Training each configuration for {n_iterations} iterations\n")

    # Test all combinations
    for hidden in hidden_layers_range:
        for neurons in neurons_range:
            result = train_and_evaluate_model(
                hidden_layers=hidden,
                neurons_per_layer=neurons,
                domain_limits=domain_limits,
                n_iterations=n_iterations
            )
            results.append(result)

            # Print intermediate result
            print(f"Model with {hidden} hidden layers, {neurons} neurons: L2 error = {result['u_l2']:.6e}, Linf error = {result['u_linf']:.6e}")

    # Sort by total error (L2 error of u + L2 error of v)
    results.sort(key=lambda x: x['total_error'])

    # Plot results
    plt.figure(figsize=(12, 8))

    # Create a matrix for heatmap
    hidden_vals = sorted(hidden_layers_range)
    neuron_vals = sorted(neurons_range)
    error_matrix = np.zeros((len(hidden_vals), len(neuron_vals)))

    for result in results:
        i = hidden_vals.index(result['hidden_layers'])
        j = neuron_vals.index(result['neurons'])
        error_matrix[i, j] = result['u_l2']

    # Plot heatmap
    plt.subplot(2, 2, 1)
    im = plt.imshow(error_matrix, cmap='viridis_r')
    plt.colorbar(im, label='L2 Error (u)')
    plt.xticks(range(len(neuron_vals)), neuron_vals)
    plt.yticks(range(len(hidden_vals)), hidden_vals)
    plt.xlabel('Neurons per Layer')
    plt.ylabel('Hidden Layers')
    plt.title('L2 Error Heatmap')

    # Plot parameter vs error
    plt.subplot(2, 2, 2)
    param_values = [r['hidden_layers'] * r['neurons'] for r in results]
    errors = [r['u_l2'] for r in results]

    # Sort by parameter value
    param_errors = sorted(zip(param_values, errors))
    param_values = [p for p, _ in param_errors]
    errors = [e for _, e in param_errors]

    plt.plot(param_values, errors, 'o-')
    plt.xlabel('Parameter Value (hidden × neurons)')
    plt.ylabel('L2 Error')
    plt.yscale('log')
    plt.grid(True)
    plt.title('Error vs Parameter Value')

    # Bar plot of errors
    plt.subplot(2, 2, 3)
    model_names = [f"{r['hidden_layers']}x{r['neurons']}" for r in results[:10]]  # Top 10
    l2_errors = [r['u_l2'] for r in results[:10]]

    y_pos = np.arange(len(model_names))
    plt.barh(y_pos, l2_errors)
    plt.yticks(y_pos, model_names)
    plt.xlabel('L2 Error')
    plt.title('Top Model Configurations')

    # Plot best vs worst model error
    plt.subplot(2, 2, 4)
    best_config = f"{results[0]['hidden_layers']}x{results[0]['neurons']}"
    worst_config = f"{results[-1]['hidden_layers']}x{results[-1]['neurons']}"

    plt.bar([0, 1], [results[0]['u_l2'], results[-1]['u_l2']])
    plt.xticks([0, 1], [f"Best: {best_config}", f"Worst: {worst_config}"])
    plt.ylabel('L2 Error')
    plt.yscale('log')
    plt.title('Best vs Worst Model')

    plt.tight_layout()
    plt.savefig('hyperparameter_tuning.png', dpi=300, bbox_inches='tight')
    plt.show()

    # Print table of results
    print("\n=== Hyperparameter Tuning Results ===")
    print(f"{'Rank':<5} | {'Hidden':<6} | {'Neurons':<7} | {'L2 Error (u)':<12} | {'Linf Error (u)':<13} | {'Total Error':<11}")
    print("-" * 65)

    for i, result in enumerate(results[:10]):  # Show top 10
        print(f"{i+1:<5} | {result['hidden_layers']:<6} | {result['neurons']:<7} | {result['u_l2']:<12.6e} | {result['u_linf']:<13.6e} | {result['total_error']:<11.6e}")

    return results[0]  # Return best model configuration


def compare_solutions(model, t_values=[0.5, 1.0, 5.0, 10.0]):
    """Compare PINN predictions with exact solutions at specific times"""
    x = torch.linspace(-math.pi, math.pi, 200).reshape(-1, 1).to(device)

    # Set up the figure
    fig, axes = plt.subplots(len(t_values), 2, figsize=(14, 4*len(t_values)))

    for i, t_val in enumerate(t_values):
        # Create time tensor
        t = torch.full_like(x, t_val).to(device)

        # Get PINN predictions
        with torch.no_grad():
            u_pred = model.u(x, t).cpu().numpy()
            v_pred = model.v(x, t).cpu().numpy()

        # Calculate exact solutions
        u_exact, v_exact = exact_solution(x.cpu(), torch.tensor([[t_val]]).cpu())
        u_exact = u_exact.numpy()
        v_exact = v_exact.numpy()

        # Plot u solution
        if len(t_values) == 1:
            ax_u = axes[0]
            ax_v = axes[1]
        else:
            ax_u = axes[i, 0]
            ax_v = axes[i, 1]

        ax_u.plot(x.cpu().numpy(), u_pred, 'r-', label='PINN')
        ax_u.plot(x.cpu().numpy(), u_exact, 'b--', label='Exact')
        ax_u.set_xlabel('x')
        ax_u.set_ylabel('u(x,t)')
        ax_u.set_title(f'u at t = {t_val}')
        ax_u.legend()
        ax_u.grid(True)

        # Plot v solution
        ax_v.plot(x.cpu().numpy(), v_pred, 'r-', label='PINN')
        ax_v.plot(x.cpu().numpy(), v_exact, 'b--', label='Exact')
        ax_v.set_xlabel('x')
        ax_v.set_ylabel('v(x,t)')
        ax_v.set_title(f'v at t = {t_val}')
        ax_v.legend()
        ax_v.grid(True)

    plt.tight_layout()
    plt.savefig('solution_comparison.png', dpi=300, bbox_inches='tight')
    plt.show()


def train_model(model, optimizer, domain_limits, n_collocation_points=10000,
               n_initial_points=200, n_boundary_points=200, epochs=10000,
               verbose=True, check_interval=500):
    """
    Train the PINN model to solve the coupled Burgers equation
    """
    x_min, x_max = domain_limits[0]
    t_min, t_max = domain_limits[1]

    # Generate collocation points for enforcing PDE
    x_collocation = torch.FloatTensor(n_collocation_points, 1).uniform_(x_min, x_max).to(device)
    t_collocation = torch.FloatTensor(n_collocation_points, 1).uniform_(t_min, t_max).to(device)

    # Generate points for initial condition at t=0
    x_initial = torch.FloatTensor(n_initial_points, 1).uniform_(x_min, x_max).to(device)
    t_initial = torch.zeros(n_initial_points, 1).to(device)

    # Generate points for boundary conditions at x=-π and x=π
    t_boundary = torch.FloatTensor(n_boundary_points, 1).uniform_(t_min, t_max).to(device)
    x_boundary_left = torch.full((n_boundary_points//2, 1), x_min).to(device)
    x_boundary_right = torch.full((n_boundary_points//2, 1), x_max).to(device)
    x_boundary = torch.cat([x_boundary_left, x_boundary_right], dim=0)
    t_boundary = torch.cat([t_boundary[:n_boundary_points//2], t_boundary[n_boundary_points//2:]], dim=0)

    # Calculate exact solutions for initial and boundary conditions
    u_exact_init, v_exact_init = exact_solution(x_initial, t_initial)
    u_exact_bound, v_exact_bound = exact_solution(x_boundary, t_boundary)

    # Training loop
    start_time = time.time()
    history = {'loss': [], 'pde_loss': [], 'ic_loss': [], 'bc_loss': [], 'time': []}

    pbar = tqdm(range(epochs)) if verbose else range(epochs)
    for epoch in pbar:
        optimizer.zero_grad()

        # PDE loss for collocation points
        f_u, f_v = model.compute_pde_residuals(x_collocation, t_collocation)
        pde_loss = torch.mean(f_u**2) + torch.mean(f_v**2)

        # Initial condition loss
        u_pred_init = model.u(x_initial, t_initial)
        v_pred_init = model.v(x_initial, t_initial)
        ic_loss = torch.mean((u_pred_init - u_exact_init)**2) + torch.mean((v_pred_init - v_exact_init)**2)

        # Boundary condition loss
        u_pred_bound = model.u(x_boundary, t_boundary)
        v_pred_bound = model.v(x_boundary, t_boundary)
        bc_loss = torch.mean((u_pred_bound - u_exact_bound)**2) + torch.mean((v_pred_bound - v_exact_bound)**2)

        # Total loss with weighting factors
        loss = pde_loss + 10 * ic_loss + 10 * bc_loss

        loss.backward()
        optimizer.step()

        # Store loss history
        if epoch % check_interval == 0:
            elapsed = time.time() - start_time
            history['loss'].append(loss.item())
            history['pde_loss'].append(pde_loss.item())
            history['ic_loss'].append(ic_loss.item())
            history['bc_loss'].append(bc_loss.item())
            history['time'].append(elapsed)

            if verbose:
                pbar.set_description(f"Loss: {loss.item():.6e}")

    # Plot loss history
    plt.figure(figsize=(12, 6))
    plt.semilogy(range(0, epochs, check_interval), history['loss'], label='Total Loss')
    plt.semilogy(range(0, epochs, check_interval), history['pde_loss'], label='PDE Loss')
    plt.semilogy(range(0, epochs, check_interval), history['ic_loss'], label='IC Loss')
    plt.semilogy(range(0, epochs, check_interval), history['bc_loss'], label='BC Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.title('Training Loss History')
    plt.savefig('loss_history.png', dpi=300, bbox_inches='tight')

    total_time = time.time() - start_time
    print(f"Training completed in {total_time:.2f} seconds")

    return model, x_collocation, t_collocation, x_initial, t_initial, x_boundary, t_boundary


def main():
    """Main function to run the PINN solution for coupled Burgers equations"""
    torch.manual_seed(42)  # For reproducibility

    # Define domain limits
    domain_limits = [(-math.pi, math.pi), (0, 10)]  # [x_domain, t_domain]

    # Define hyperparameter search space
    hidden_layers_range = [3, 4, 5, 6, 7]  # Range of hidden layers to test
    neurons_range = [20, 30, 40, 50, 60]    # Range of neurons per layer to test

    print("Step 1: Performing hyperparameter tuning")
    best_config = hyperparameter_tuning(
        domain_limits,
        hidden_layers_range,
        neurons_range,
        n_iterations=1000  # Quick training for tuning
    )

    # Extract best hyperparameters
    hidden_layers = best_config['hidden_layers']
    neurons = best_config['neurons']

    print(f"\nStep 2: Training full model with best hyperparameters (hidden_layers={hidden_layers}, neurons={neurons})")
    # Initialize the model with best hyperparameters
    model = PINN(hidden_layers=hidden_layers, neurons_per_layer=neurons).to(device)
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    # Print model architecture
    print(model)

    # Train the model with more epochs
    print("Training optimized model...")
    model, x_col, t_col, x_init, t_init, x_bound, t_bound = train_model(
        model, optimizer, domain_limits, epochs=5000, check_interval=500
    )

    # Evaluate model performance at specific times
    print("\nCalculating errors for specific time points...")
    x_domain = torch.linspace(-math.pi, math.pi, 200).reshape(-1, 1).to(device)
    errors = compute_errors(model, x_domain, [0.5, 1.0, 5.0, 10.0])

    # Print error metrics in a table format
    print("\nError metrics for the optimized model:")
    print("=" * 60)
    print(f"{'t':<5} | {'u L∞':<15} | {'u L2':<15} | {'v L∞':<15} | {'v L2':<15}")
    print("-" * 60)
    for t_val, err in errors.items():
        print(f"{t_val:<5} | {err['u_linf']:<15.6e} | {err['u_l2']:<15.6e} | {err['v_linf']:<15.6e} | {err['v_l2']:<15.6e}")
    print("=" * 60)

    # Create plots
    print("\nGenerating solution plots...")
    compare_solutions(model)
    plot_3d_solutions(model, domain_limits)

    print("\nAnalysis complete!")


if __name__ == "__main__":
    main()
```
